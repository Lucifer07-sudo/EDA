## Clean dataset

### Drop unnecessary columns

```python
df = df.drop(['url', 'address', 'phone', 'menu_item', 'reviews_list' ], axis = 1)
df.head()
```

```python
#info about df
df.info()
```

```python
#checking for null values
round(df.isna().sum()/len(df)*100, 2)
```

```python
#dropping dish_liked column since it has more than 50% null values
df = df.drop(['dish_liked'], axis=1)
df.head()
```

```python
#removing duplicate values
df.drop_duplicates(inplace=True)
df.shape
```

### Clean Rate column
- Cleaning to be done:
- Remove '/5', keep only float vlaue
- Remove any invalid string or value
- Fill null values with mean value

```python
df.rate.unique()
```

```python
#function to remove /5 and remove string or other invalid values like '-'

def handlerate(val):
  if val == 'NEW' or val == '-':
    return np.nan
  else:
    val = str(val).split('/')
    val = val[0]
    return float(val)

df.rate = df.rate.apply(handlerate)
df.rate.unique()
```

```python
df.rate.isnull().sum()
```

```python
#fill null values with mean of rate column
df.rate.fillna(df.rate.mean(), inplace=True)
df.rate.isnull().sum()
```

```python
round(df.isnull().sum()/len(df)*100,2)
```

#### Since columns location, rest_type, cuisines, approx_cost(for two people) have very less null values, we will directly drop those rows                      

```python
df.dropna(inplace=True)
df.shape
```

```python
round(df.isnull().sum()/len(df)*100,2)
```

#### Renaming column approx_cost(for two people), listed_in(type), listed_in(city) for ease of use

```python
df.rename(columns={'approx_cost(for two people)':'Cost2plates', 'listed_in(type)':'type', 'listed_in(city)':'city'}, inplace=True)
df.head()
```

#### Both the city and location columns seem to have the same information i.e, restraunt location, we can drop one of these, in this case I will drop the city column

```python
df =df.drop(['city'], axis=1)
df.head()
```

```python
#checkout the Cost2plates column
df.Cost2plates.unique()
```

#### Remove the commas and convert values to floats since these are cost values

```python
#function to remove commas
def handleCost2plates(value):
  if ',' in value:
    value = value.replace(',','')
    return float(value)
  else:
    return float(value)

df.Cost2plates = df.Cost2plates.apply(handleCost2plates)
df.Cost2plates.unique()
```

### Changing rest_type column

```python
#checking unique value counts
df.rest_type.value_counts()
```

```python
#Creating a Series where the index is the unique values in the rest_type column and the values are the counts of each unique value
rest_types = df.rest_type.value_counts(ascending = False)
rest_types
```

```python
#filtering out types which have less than 1000 restraunts
rest_types_lessthan1000 = rest_types[rest_types<1000]
rest_types_lessthan1000
```

```python
#function to club all restraunt types with less than 1000 restraunt count into a new type 'others
def handleRestType(value):
  if value in rest_types_lessthan1000:
    return 'others'
  else:
    return value

df.rest_type = df.rest_type.apply(handleRestType)
df.rest_type.value_counts()
```

### Cleaning location column

```python
df.location.value_counts()
```

```python
locations = df.location.value_counts(ascending = False)
locations
```

```python
locationLessthan_300 = locations[locations<300]
locationLessthan_300
```

```python
def handleLocation(value):
  if value in locationLessthan_300:
    return 'others'
  else:
    return value

df.location = df.location.apply(handleLocation)
df.location.value_counts()
```

```python
df.head()
```

### Cuisine column

```python
df.cuisines.value_counts()
```

```python
cuisine = df.cuisines.value_counts(ascending=False)
cuisine
```

```python
cuisinelessthan_100 = cuisine[cuisine<100]
cuisinelessthan_100
```

```python
def handleCuisine(value):
  if value in cuisinelessthan_100:
    return 'others'
  else:
    return value

df.cuisines = df.cuisines.apply(handleCuisine)
df.cuisines.value_counts()
```
